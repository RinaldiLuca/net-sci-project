```python
#Here we can build the matrix

network_matrix = np.zeros((len(country_iso),len(country_iso)))
for i in range(0,len(country_iso)):
    temporal_df = test[(test["rt3ISO"]==country_iso[i])]
    
    for j in range(0,len(country_iso)):
        temporal_df2 = temporal_df[temporal_df["pt3ISO"]==country_iso[j]]
        if temporal_df2.empty:
            network_matrix[i,j] = 0
        else:
            network_matrix[i,j] = temporal_df2["TradeValue"].values[0]

        
```

```python
# a 2D array with linearly increasing values on the diagonal
plt.figure(figsize=(10,10))
plt.matshow(network_matrix, fignum=1)

plt.show()
```

```python
country_iso.index("KAZ")
```

```python
count_node_a = []
count_node_b = []

for i in range(0,len(country_iso)):
    count_node_a.append(np.count_nonzero(network_matrix[i,]))
    count_node_b.append(np.count_nonzero(network_matrix[:,i]))
```

```python
# the histogram of the data
plt.figure(figsize=(7,5))
n, bins, patches = plt.hist(count_node_a, 30, density=True, alpha=0.75)
n, bins, patches = plt.hist(count_node_b, 30, density=True, alpha=0.75)


plt.title('Degree distribution')
plt.ylabel(r'$p_k$')
plt.xlabel('k')
plt.grid(True)
plt.show()
```
